```python
app_metadata_df = pd.read_parquet(processed_dir / "app_metadata.parquet")
app_category_counts = (
    app_metadata_df["app_category"]
    .fillna("<missing>")
    .value_counts(dropna=False)
    .rename_axis("app_category")
    .reset_index(name="count")
    .sort_values(["count", "app_category"], ascending=[False, True], kind="stable")
    .reset_index(drop=True)
)
print(f"Unique categories: {app_metadata_df['app_category'].nunique(dropna=True)}")
app_category_mapping_df = app_category_counts.copy()
app_category_mapping_df.insert(0, "app_category_id", range(1, len(app_category_mapping_df) + 1))
app_category_mapping_path = processed_dir / "app_category_mapping.parquet"
app_category_mapping_df.to_parquet(app_category_mapping_path, index=False)
print(f"Saved category mapping to {app_category_mapping_path}")
app_category_mapping_df
```

```python
# Smoke test: set SEQUENCE_SAMPLE_N to a small number like 1000.
# Full execution: set SEQUENCE_SAMPLE_N = None.
# Use "head" for a fast deterministic check, or "sample" for a random slice.
SEQUENCE_SAMPLE_N = None
SEQUENCE_SAMPLE_MODE = "head"

sequences_df = pd.read_parquet(processed_dir / "sequences.parquet")
if SEQUENCE_SAMPLE_N is not None:
    if SEQUENCE_SAMPLE_MODE == "sample":
        sequences_df = sequences_df.sample(
            n=min(SEQUENCE_SAMPLE_N, len(sequences_df)),
            random_state=42,
        ).reset_index(drop=True)
    else:
        sequences_df = sequences_df.head(SEQUENCE_SAMPLE_N).copy()
print(f"Running enrichment on {len(sequences_df):,} sequence rows")

item_mapping_df = pd.read_parquet(processed_dir / "item_mapping.parquet")
app_metadata_df = pd.read_parquet(processed_dir / "app_metadata.parquet")
app_category_mapping_df = pd.read_parquet(processed_dir / "app_category_mapping.parquet")

item_features_df = (
    item_mapping_df
    .merge(
        app_metadata_df[["app_package", "app_category", "num_reviews", "avg_rating", "price"]],
        on="app_package",
        how="left",
    )
    .merge(
        app_category_mapping_df[["app_category_id", "app_category"]],
        on="app_category",
        how="left",
    )
)

item_features_df["app_category_id"] = item_features_df["app_category_id"].fillna(0).astype(int)
item_features_df["num_reviews_numeric"] = pd.to_numeric(
    item_features_df["num_reviews"].astype("string").str.replace(",", "", regex=False),
    errors="coerce",
).fillna(0).astype(int)
item_features_df["avg_rating_scaled"] = (
    ((item_features_df["avg_rating"].fillna(0) - 1.0) / 4.0)
    .clip(lower=0.0, upper=1.0)
    .round(4)
)
item_features_df["price_numeric"] = pd.to_numeric(
    item_features_df["price"].astype("string").str.extract(r"\$\s*([0-9]+(?:\.[0-9]+)?)", expand=False),
    errors="coerce",
).fillna(0.0).round(2)

category_lookup = item_features_df.set_index("item_id")["app_category_id"].to_dict()
num_reviews_lookup = item_features_df.set_index("item_id")["num_reviews_numeric"].to_dict()
avg_rating_lookup = item_features_df.set_index("item_id")["avg_rating_scaled"].to_dict()
price_lookup = item_features_df.set_index("item_id")["price_numeric"].to_dict()

sequences_enriched_df = (
    sequences_df.drop(columns=["app_packages"])
    .rename(columns={"votes": "review_upvotes"})
    .copy()
)
# Preserve the original per-user, per-item vote counts under an explicit feature name.
sequences_enriched_df["review_upvotes"] = sequences_enriched_df["review_upvotes"].apply(
    lambda seq: [int(vote) for vote in seq]
)
# Binarize the original 1-5 ratings into implicit feedback labels.
sequences_enriched_df["ratings"] = sequences_enriched_df["ratings"].apply(
    lambda seq: [1 if float(rating) >= 3 else -1 for rating in seq]
)
sequences_enriched_df["app_category"] = sequences_enriched_df["test_sequence"].apply(
    lambda seq: [[int(category_lookup.get(item_id, 0))] for item_id in seq]
)
sequences_enriched_df["app_num_reviews"] = sequences_enriched_df["test_sequence"].apply(
    lambda seq: [int(num_reviews_lookup.get(item_id, 0)) for item_id in seq]
)
sequences_enriched_df["app_avg_rating"] = sequences_enriched_df["test_sequence"].apply(
    lambda seq: [float(avg_rating_lookup.get(item_id, 0.0)) for item_id in seq]
)
sequences_enriched_df["app_price"] = sequences_enriched_df["test_sequence"].apply(
    lambda seq: [float(price_lookup.get(item_id, 0.0)) for item_id in seq]
)

sequences_enriched_df.head()
```

```python
interactions_df = pd.read_parquet(processed_dir / "interactions.parquet")
app_category_mapping_df = pd.read_parquet(processed_dir / "app_category_mapping.parquet")

interaction_category_lookup = app_category_mapping_df.set_index("app_category")["app_category_id"].to_dict()
interactions_enriched_df = (
    interactions_df.drop(columns=["app_package", "uid", "app_category"])
    .rename(columns={"votes": "review_upvotes"})
    .copy()
)
interactions_enriched_df["review_upvotes"] = interactions_enriched_df["review_upvotes"].fillna(0).astype(int)
interactions_enriched_df["rating"] = interactions_enriched_df["rating"].apply(
    lambda rating: 1 if float(rating) >= 3 else -1
)
interactions_enriched_df["app_category"] = (
    interactions_df["app_category"]
    .map(interaction_category_lookup)
    .fillna(0)
    .astype(int)
    .apply(lambda x: [x])
)

enriched_interactions_path = processed_dir / "enriched_mobilerec_interactions.parquet"
interactions_enriched_df.to_parquet(enriched_interactions_path, index=False)
print(f"Saved enriched interactions to {enriched_interactions_path}")
interactions_enriched_df
```

```python
# Smoke test: keep UPLOAD_TO_KAGGLE = False.
# Full publish: set UPLOAD_TO_KAGGLE = True and keep SEQUENCE_SAMPLE_N = None in the previous cell.
UPLOAD_TO_KAGGLE = SEQUENCE_SAMPLE_N is None

from models.upload_processed_to_kaggle import (
    collect_dataset_files,
    load_kaggle_credentials,
    upload_kaggle_dataset_snapshot,
)

enriched_sequences_path = processed_dir / "enriched_mobilerec_sequences.parquet"
enriched_interactions_path = processed_dir / "enriched_mobilerec_interactions.parquet"
sequences_enriched_df.to_parquet(enriched_sequences_path, index=False)
print(f"Saved enriched sequences to {enriched_sequences_path}")
if not enriched_interactions_path.exists():
    raise FileNotFoundError(
        f"Run the interactions enrichment cell first: {enriched_interactions_path}"
    )

if UPLOAD_TO_KAGGLE:
    env_path, kaggle_username, _ = load_kaggle_credentials(Path(".env"))
    kaggle_dataset_handle = f"{kaggle_username}/cs5242-mobilerec-dataset"
    dataset_files = collect_dataset_files(processed_dir)
    dataset_files.pop("enriched_mobilerec_sequences.parquet", None)
    dataset_files.pop("enriched_mobilerec_interactions.parquet", None)
    dataset_files["enriched_mobilerec_sequences.parquet"] = enriched_sequences_path.resolve()
    dataset_files["enriched_mobilerec_interactions.parquet"] = enriched_interactions_path.resolve()
    upload_kaggle_dataset_snapshot(
        kaggle_dataset_handle,
        dataset_files,
        version_notes="refresh mobilerec dataset with enriched mobilerec sequences and interactions parquet",
    )
else:
    print("Skipped Kaggle upload. Set UPLOAD_TO_KAGGLE = True for the full publish run.")
```
